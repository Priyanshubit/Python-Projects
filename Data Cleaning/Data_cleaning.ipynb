# 3. Cleaning and Formatting Phone_Number

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)  
```

```python
df
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))      #Converting number into str
```

```python
df
```

```python
df['Phone_Number'].apply(lambda x: x[0:3]+'-' + x[3:6] + '-' +x[6:10] + '-')      #setting a 3 number format
```

```python
df
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('nan---','',regex=True)
```

```python
df
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('nan','NaN',regex=True)
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('NaNN','NaN',regex=True)
```

```python
df
```

# 4. Splitting and Arranging the Address Column

```python
df[['Street address','State','Zipcode']]=df['Address'].str.split(',',n=2, expand=True)
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Y','Yes')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('N','No')
df
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('Y','Yes')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df['Paying Customer']=df['Paying Customer'].str.replace('N','No')
df
```
