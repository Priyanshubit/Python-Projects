# Pandas Visualisation

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\acer\Downloads\Python\Pandas\Ice Cream Ratings.csv")
```

```python
df=df.set_index('Date')
```

```python
df
```

```python
# To visualise something in pandas we use plot
```

```python
df.plot()
```

# Line Chart

```python
df.plot(kind='line',subplots=True)
```

```python
df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Scores')
```

# Bar Chart

```python
df.plot(kind='bar', title='Ice cream Ratings', xlabel='Monthly Ratings',ylabel='Scores')
```

# Stacked Bar Chart

```python
df.plot(kind='bar', stacked=True, title='Ice cream Ratings', xlabel='Monthly Ratings',ylabel='Scores')
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True, title='Ice cream Ratings', xlabel='Monthly Ratings',ylabel='Scores')
```

```python
df.plot.barh(stacked=True, title='Ice cream Ratings', xlabel='Monthly Ratings',ylabel='Scores')
```

# Scatter Plot Chart

```python
df.plot.scatter(x='Texture Rating',y='Overall Rating', s=500, c='Yellow')
```

# Histogram

```python
df.plot.hist(bins=20)
```

# Box Plot

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5,))
```

# Pie Chart

```python
df.plot.pie(y='Flavor Rating',figsize=(4,4))
```
