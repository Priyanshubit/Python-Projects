```python
# Average Population Growth by Continent (1970–2022) with Mean Land Area
```

```python
df3=df.groupby('Continent')[df.columns[13:4:-1]].mean(numeric_only=True).sort_values(by='2022 Population',ascending=False)
```

```python
df3
```

```python
# Which continent has the highest average population till now
```

```python
df2 = (
    df.groupby('Continent')[df.columns[13:4:-1]]          # same column slice
      .mean(numeric_only=True)                         # your per‑continent means
      .assign(Total=lambda x: x.sum(axis=1, numeric_only=True))  # new sum column
      .sort_values(by='2022 Population', ascending=False)        # keep the sort
)
```

```python
df2
```

```python
# Transpose the data and visulaise it in plot 
```

```python
df4=df2.transpose()
```

```python
df4
```
