# Group by

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\acer\Downloads\Python\Pandas\Flavors.csv") 
df                                         #Import csv file
```

```python
group_by_frame=df.groupby('Base Flavor')     # Group by uisng Base flavour
```

```python
group_by_frame.mean(numeric_only=True)    #use numeric only as it is not able to read integer
```

```python
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').count()        #COUNT
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby(['Base Flavor', 'Texture Rating']).mean(numeric_only=True)
```

```python
df.groupby(['Base Flavor','Liked']).mean(numeric_only=True)
```

# AGGREGATE FUNCTIONS

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','sum','max','min','count'],'Texture Rating':['mean','sum','max','min','count']})
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','sum','max','min','count'],'Texture Rating':['mean','sum','max','min','count']})
```
