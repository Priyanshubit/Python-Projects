# MERGE JOIN AND CONCATENATE

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\acer\Downloads\Python\Pandas\LOTR.csv")
```

```python
df
```

```python
df1=pd.read_csv(r"C:\Users\acer\Downloads\Python\Pandas\LOTR 2.csv")
```

```python
df1
```

# MERGE

```python
df.merge(df1)
```

# INNER JOIN

```python
df.merge(df1, how='inner', on=['FellowshipID'])
```

```python
df.merge(df1, how='inner', on=['FellowshipID','FirstName'])
```

# OUTER JOIN

```python
df.merge(df1, how='outer')
```

# LEFT JOIN

```python
df.merge(df1, how='left')
```

# RIGHT JOIN

```python
df.merge(df1, how='right')
```

# CROSS JOIN

```python
df.merge(df1, how='cross')
```

# JOIN

```python
df.join(df1, on='FellowshipID', how='outer', lsuffix='_Left_data', rsuffix='_Right_data')
```

```python
df2=df.set_index('FellowshipID').join(df1.set_index('FellowshipID'), lsuffix='_Left', rsuffix='_Right', how='outer')
```

```python
df2
```

# CONCATENATE

```python
pd.concat([df,df1])
```

```python
pd.concat([df,df1], join='outer')
```

```python
pd.concat([df,df1], join='inner')
```

```python
pd.concat([df,df1], join='outer', axis=1)    # JUST LIKE MERGE SIDE BY SIDE
```
